```python
def path_creation(Industry,path):
    os.chdir(Path)
    List=[]
    # iterate through all file
    for file in os.listdir():
    # Check whether file is in text format or not
        if file.endswith(".txt"):
            file_path = f"{Path}{file}"
            # call read text file function
            string_text_file(file_path,List,Industry)
    
    return (List)

def string_text_file(file_path,l,string1):
    with open(file_path, encoding='cp1252') as file1: 
# Loop through the file line by line
        for line in file1: 
            index =+1
            if string1 in line: 
                l.append(file1.name)
def txtcsv (directorylist):
    import re
    l2=[]
    l3=[]
    for i in range (len(directorylist)):
        regx=re.compile('txt')
        l2.append(regx.sub('csv',directorylist[i]))
        l3.append(l2[i][-17:])
    return l2

def namecsv(l):
    l3=[]
    for i in range (0,len(l)):
        l3.append(l[i][-17:])
    return l3

def readcsv(directory):
    df3=pd.read_csv(directory, usecols= ['Date','Time','Ele'])
    l=[]
    for i in range (0,len(df3)):
        l.append(df3['Date'][i][5:])

    df3.insert(loc=1, column='Month',value=np.array(l))

    return(df3)

def month (df3):
    l=[]
    for i in range (0,df3.shape[0]):
        l.append(df3['Date'][i][5:])
    df3.insert(loc=1, column='Month',value=np.array(l))

    return(df3)    
def date_extracter (directory):
    df=readcsv(directory)
    date_list=df.at[0,'Date'], df.at[len(df)-1,'Date']
    return (date_list)

def consumption_extracter (directory):
    df=readcsv(directory)
    l=[df['Ele'].mean(), df['Ele'].max()]
    return (l)

def average_con2(df):
    l=[]
    lwi=[]
    lsp=[]
    lsu=[]
    lau=[]
    for i in range (1,48):
        l.append((i+1)/2)
        lwi.append((df.loc[((df['Month'] >= '12-01') | (df['Month'] <= '02-28')) & (df['Time'] == i )]['Ele']).mean())
        lsp.append((df.loc[((df['Month'] >= '03-01') & (df['Month'] <= '05-31')) & (df['Time'] == i )]['Ele']).mean())
        lsu.append((df.loc[((df['Month'] >= '06-01') & (df['Month'] <= '08-31')) & (df['Time'] == i )]['Ele']).mean())
        lau.append((df.loc[((df['Month'] >= '09-01') & (df['Month'] <= '11-30')) & (df['Time'] == i )]['Ele']).mean())
    
    df1 = pd.DataFrame({"Time": l, "Winter Consumption": lwi, "Spring Consumption": lsp, "Summer Consumption": lsu, "Autumn Consumption": lau})
    return df1

def general_analysis (Industry,Path):
#List of csv files directories
    List=path_creation(Industry,Path)
    data_list=txtcsv(List)
    names=namecsv(txtcsv(List))#List of the names of csv files
    dfwi=pd.DataFrame({namecsv(data_list)[0]:average_con2(data_list[0])['Winter Consumption'].tolist()})
    dfsp=pd.DataFrame({namecsv(data_list)[0]:average_con2(data_list[0])['Spring Consumption'].tolist()})
    dfsu=pd.DataFrame({namecsv(data_list)[0]:average_con2(data_list[0])['Summer Consumption'].tolist()})
    dfau=pd.DataFrame({namecsv(data_list)[0]:average_con2(data_list[0])['Autumn Consumption'].tolist()})
    for i in range (1,len(data_list)):
        average_con2(data_list[i])['Summer Consumption'].tolist()
        namecsv(data_list)[i]
        dfwi.insert(loc=i,column=namecsv(data_list)[i],value=np.array(average_con2(data_list[i])['Winter Consumption'].tolist()))
        dfsp.insert(loc=i,column=namecsv(data_list)[i],value=np.array(average_con2(data_list[i])['Spring Consumption'].tolist()))
        dfsu.insert(loc=i,column=namecsv(data_list)[i],value=np.array(average_con2(data_list[i])['Summer Consumption'].tolist()))
        dfau.insert(loc=i,column=namecsv(data_list)[i],value=np.array(average_con2(data_list[i])['Autumn Consumption'].tolist()))   
    df_final=pd.DataFrame({ "Time": average_con2(data_list[0])['Time'] , "Average Winter": np.array(dfwi.sum(axis = 1, skipna = True)/(len(dfwi.columns)))})
    df_final.insert(loc=2, column= "Average Spring", value=np.array(dfsp.sum(axis = 1, skipna = True)/(len(dfsp.columns))))
    df_final.insert(loc=3, column= "Average Summer", value=np.array(dfsu.sum(axis = 1, skipna = True)/(len(dfsu.columns))))
    df_final.insert(loc=4, column= "Average Autumn", value=np.array(dfau.sum(axis = 1, skipna = True)/(len(dfau.columns))))
    return (df_final)

def seasonal_plotting(df_final,Industry):
    plt.plot( 'Time', "Winter Consumption", data=df_final, marker='o', markerfacecolor='grey', markersize=3, color='grey', linewidth=3)
    plt.plot( 'Time', "Spring Consumption", data=df_final, marker='o', markerfacecolor='green', markersize=3, color='green', linewidth=3)
    plt.plot( 'Time', "Summer Consumption", data=df_final, marker='o', markerfacecolor='yellow', markersize=3, color='yellow', linewidth=3)
    plt.plot( 'Time', "Autumn Consumption", data=df_final, marker='o', markerfacecolor='orange', markersize=3, color='orange', linewidth=3)
# show legend
    plt.legend(loc="lower center")
    plt.ylabel("Electric consumption [kW]")
    plt.xlabel("Hours")
    plt.title(Industry)
    plt.locator_params(axis="x", nbins=24)
    plt.xlim(0, 25)
# show graph
    plt.show()
    

    
def industrial_park(df,Path):
    dwi=pd.DataFrame({ "Time": [1.0,1.5,2.0, 2.5,3.0,3.5,4.0,4.5,5.0,5.5,6.0,6.5,7.0,7.5,8.0,8.5,9.0,9.5,10.0,10.5,11.0,11.5,12.0,12.5,13.0,13.5,14.0,14.5,15.0,15.5,16.0,16.5,17.0,17.5,18.0,18.5,19.0,19.5,20.0,20.5,21.0,21.5,22.0,22.5,23.0,23.5,24.0]})
    dsp=pd.DataFrame({ "Time": [1.0,1.5,2.0, 2.5,3.0,3.5,4.0,4.5,5.0,5.5,6.0,6.5,7.0,7.5,8.0,8.5,9.0,9.5,10.0,10.5,11.0,11.5,12.0,12.5,13.0,13.5,14.0,14.5,15.0,15.5,16.0,16.5,17.0,17.5,18.0,18.5,19.0,19.5,20.0,20.5,21.0,21.5,22.0,22.5,23.0,23.5,24.0]})
    dsu=pd.DataFrame({ "Time": [1.0,1.5,2.0, 2.5,3.0,3.5,4.0,4.5,5.0,5.5,6.0,6.5,7.0,7.5,8.0,8.5,9.0,9.5,10.0,10.5,11.0,11.5,12.0,12.5,13.0,13.5,14.0,14.5,15.0,15.5,16.0,16.5,17.0,17.5,18.0,18.5,19.0,19.5,20.0,20.5,21.0,21.5,22.0,22.5,23.0,23.5,24.0]})
    dau=pd.DataFrame({ "Time": [1.0,1.5,2.0, 2.5,3.0,3.5,4.0,4.5,5.0,5.5,6.0,6.5,7.0,7.5,8.0,8.5,9.0,9.5,10.0,10.5,11.0,11.5,12.0,12.5,13.0,13.5,14.0,14.5,15.0,15.5,16.0,16.5,17.0,17.5,18.0,18.5,19.0,19.5,20.0,20.5,21.0,21.5,22.0,22.5,23.0,23.5,24.0]})
    df_industrialpark=pd.DataFrame({ "Time": [1.0,1.5,2.0, 2.5,3.0,3.5,4.0,4.5,5.0,5.5,6.0,6.5,7.0,7.5,8.0,8.5,9.0,9.5,10.0,10.5,11.0,11.5,12.0,12.5,13.0,13.5,14.0,14.5,15.0,15.5,16.0,16.5,17.0,17.5,18.0,18.5,19.0,19.5,20.0,20.5,21.0,21.5,22.0,22.5,23.0,23.5,24.0]})
    df_result=pd.DataFrame({ "Name":[],"Max":[],"Mean":[]})
    h=1
    for i in range (0, len(df['Industry'])):
        if (df['Number'][i] >0):
            List=path_creation(df['Industry'][i],Path)
            data_list=txtcsv(List)
            names=namecsv(data_list)
            df_indus=pd.DataFrame({"Path":[] ,"Name":[], "Max": [],"Mean":[]})
            min_indus = df.at[i,'Min']
            max_indus = df.at[i,'Max']
            for j in range (0,len(data_list)):
                if (consumption_extracter(data_list[j])[1]>=min_indus) & (consumption_extracter(data_list[j])[0]<=max_indus):
                    df2=pd.DataFrame({"Path":[data_list[j]] , "Name":[names[j]],"Max": [consumption_extracter(data_list[j])[0]],"Mean":[consumption_extracter(data_list[j])[1]]})
                    df_indus=df_indus.append(df2, ignore_index= True )
            
            for g in range (0,df['Number'][i]):
                k=random.randint(0,len(df_indus))
                df3=pd.DataFrame({ "Name":[(df.at[i,'Industry']) + (df_indus.at[k, "Name"])],"Max":[consumption_extracter(df_indus["Path"][k])[1]], "Mean":[consumption_extracter(df_indus["Path"][k])[0]] })
                df_result=df_result.append(df3,ignore_index=True)
                dwi.insert(loc=h, column=(df.at[i,'Industry']) + (df_indus.at[k, "Name"]),value=np.array(average_con2(df_indus.at[k,"Path"])["Winter Consumption"].tolist()))
                dsp.insert(loc=h, column=(df.at[i,'Industry']) + (df_indus.at[k, "Name"]),value=np.array(average_con2(df_indus.at[k,"Path"])["Spring Consumption"].tolist()))
                dsu.insert(loc=h, column=(df.at[i,'Industry']) + (df_indus.at[k, "Name"]),value=np.array(average_con2(df_indus.at[k,"Path"])["Summer Consumption"].tolist()))
                dau.insert(loc=h, column=(df.at[i,'Industry']) + (df_indus.at[k, "Name"]),value=np.array(average_con2(df_indus.at[k,"Path"])["Autumn Consumption"].tolist()))
                h+=1
                #df_indus=df_indus.drop([k],axis=0)
            List=[]
            data_list=[]
            names=[]
        else:
            continue
    
    dwi=dwi.drop(columns=['Time'])
    dsu=dsu.drop(columns=['Time'])
    dsp=dsp.drop(columns=['Time'])
    dau=dau.drop(columns=['Time'])
    df_industrialpark.insert(loc=1, column="Average Winter",value=np.array(dwi.mean(axis=1,skipna=True).tolist()))
    df_industrialpark.insert(loc=2, column="Average Spring",value=np.array(dsp.mean(axis=1,skipna=True).tolist()))
    df_industrialpark.insert(loc=3, column="Average Summer",value=np.array(dsu.mean(axis=1,skipna=True).tolist()))
    df_industrialpark.insert(loc=4, column="Average Autumn",value=np.array(dau.mean(axis=1,skipna=True).tolist()))
    print(df_industrialpark)
    return(df_industrialpark)
    
def timeline (start, end,n):
    df=pd.DataFrame({"Date":[], "Time": []})
    start_date = date(int(start.split("-")[0]),int(start.split("-")[1]),int(start.split("-")[2]))
    end_date = date(int(end.split("-")[0]),int(end.split("-")[1]),int(end.split("-")[2]))
    delta = timedelta(days=1)
    while start_date <= end_date:
        for k in range (0,n+1):
            df2=pd.DataFrame({"Date":[start_date], "Time": [k/2]})
            df=df.append(df2, ignore_index= True )
        start_date+=delta
    return (df)

def datetransf(date1):
    return(date(int(date1.split("-")[0]),int(date1.split("-")[1]),int(date1.split("-")[2])))
```
